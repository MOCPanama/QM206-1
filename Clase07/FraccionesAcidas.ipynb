```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fraccion(pH,pKa):

    H=10**(-pH)
    Ka=[1]
    for pk in pKa:
        x=10**(-pk)
        Ka.append(x)
    n=len(Ka)-1
    mult=1
    terminos=[]
    for a in range(n+1):
        n1=(n-a)
        cH=H**n1
        mult=mult*Ka[a]
        d=cH*mult
        terminos.append(d)
    sumaTerminos=sum(terminos)
    alpha=terminos/sumaTerminos  
    return alpha
```

```python
pKa=np.array([4.0,6.7,10.6])
pH=0
x=[]
y=np.zeros(len(pKa)+1)

while pH<=14:
    x.append(pH)
    alpha=fraccion(pH,pKa)
    y=np.vstack((y, alpha))

    
    pH+=0.1
y=np.delete(y, 0, 0)
plt.plot(x,y)

plt.show()
```
